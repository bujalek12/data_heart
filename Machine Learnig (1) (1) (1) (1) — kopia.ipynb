##KNN

```python
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
```

```python
X, y = heart.drop(columns="AHD"), heart["AHD"]
```

```python
knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(X, y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test = train_test_split(heart,
                                   test_size=0.2, 
                                   random_state=2020)
```

```python
grid_cv = GridSearchCV(KNeighborsClassifier(), 
                       param_grid={
                           "n_neighbors": range(1, 100)
                       }, cv=5, scoring="f1", 
                       verbose=1, n_jobs=6)
```

```python
grid_cv.fit(X_train.drop(columns="AHD"), 
            X_train["AHD"])
```

```python
grid_cv.best_score_
```

```python
knn = grid_cv.best_estimator_
knn
```

```python
from sklearn.metrics import f1_score
```

```python
f1_score(X_test["AHD"],
         knn.predict(X_test.drop(columns="AHD")))
```

```python
##LogisticRegression
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_validate
```

```python
X = heart.drop(columns="AHD")
y = heart["AHD"]
```

```python
lr = LogisticRegression(penalty="none", 
                        max_iter=10000)
scores = cross_validate(
    lr, X, y, cv=10, return_train_score=True,
    scoring=("accuracy", "precision", "recall", "f1")
)
scores
```

```python
from sklearn.feature_selection import RFECV
```

```python
selector = RFECV(lr, step=1, min_features_to_select=1, 
                 cv=5, scoring="f1")
selector.fit(X, y)
```

```python
selected_features_df = pd.DataFrame({
    "selected": selector.support_,
    "ranking": selector.ranking_
}, index=X.columns)
selected_features_df
```

```python
feature_names = selected_features_df[
    selected_features_df["selected"] == True
].index
feature_names
```
